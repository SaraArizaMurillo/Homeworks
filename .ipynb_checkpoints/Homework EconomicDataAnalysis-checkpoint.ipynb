## 1) Evolution of total population across the world

```python
# Let's import pandas and some other basic packages we will need
from __future__ import division
%pylab --no-import-all
%matplotlib inline
import pandas as pd
import numpy as np
import os

pathout = './data/' #the directory of the data

if not os.path.exists(pathout):
    os.mkdir(pathout) 
    
pathgraphs = './graphs/' #the directory of the graphs
if not os.path.exists(pathgraphs):
    os.mkdir(pathgraphs)
```

### Download New Maddison Project Data

```python

try:
    maddison_new = pd.read_stata(pathout + 'Maddison2018.dta') #intente crear el paquete madison new y le pido a pandas que lea un archivo stata
    maddison_new_region = pd.read_stata(pathout + 'Maddison2018_region.dta')
    maddison_new_1990 = pd.read_stata(pathout + 'Maddison2018_1990.dta')
except:
    maddison_new = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018.dta') # lee y baja el archivo
    maddison_new.to_stata(pathout + 'Maddison2018.dta', write_index=False, version=117)
    maddison_new_region = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018_region_data.dta')
    maddison_new_region.to_stata(pathout + 'Maddison2018_region.dta', write_index=False, version=117)
    maddison_new_1990 = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018_1990bm.dta')
    maddison_new_1990.to_stata(pathout + 'Maddison2018_1990.dta', write_index=False, version=117)
```

```python
maddison_new_region #database where I have the data for the world
maddison_new_region['year'] = maddison_new_region.year.astype(int) #lets fix the variable year
maddison_new_region['pop']=maddison_new_region['pop'].apply(lambda x:x*1000) #to have the population in billions
maddison_new_region
```

### Data transformation and plotting

```python
import matplotlib as mpl
import seaborn as sns
# Setup seaborn
sns.set()
```

```python
BASE_world = maddison_new_region.loc[maddison_new_region.region_name=='World']#we create a new data base with the same data from maddison new region 
BASE_world=BASE_world.pivot_table(index='year',columns='region_name',values='pop',aggfunc='sum')
BASE_world
```

```python
#plot with pandas
plt.plot(BASE_world, color='blue', marker='o', linestyle='dashed', linewidth=1, markersize=3)
plt.title('The evolution of total population across the world')
plt.xlabel('year')
plt.ylabel('Population in billions')
plt.show()
```

```python
#plot with seaborn
sns.set(rc={'figure.figsize':(11.7,8.27)})
fig, ax = plt.subplots()
sns.lineplot( data=BASE_world, alpha=1, style='Region', dashes=False, markers=True,linewidth=2, markersize=6)
ax.tick_params(axis = 'both', which = 'major', labelsize=12)
ax.tick_params(axis = 'both', which = 'minor', labelsize=12)
ax.set_xlabel('Year')
ax.set_ylabel('Population')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title("Evolution of total population across the world", fontsize=18)
plt.savefig(pathgraphs + 'population_world.pdf', dpi=300, bbox_inches='tight')
```

## 2) Evolution of the share of the world population by countries and WB regions

```python
maddison_new['year'] = maddison_new.year.astype(int) #lets fix the variable year
maddison_new
```

```python
BASE_country=maddison_new 
BASE_tot=BASE_country.merge(BASE_world, on=["year"])
BASE_tot['pop']=BASE_tot['pop'].apply(lambda x:x*1000) #to have the population in billions
BASE_tot
```

```python
#share of the world population by countries
BASE_tot1=BASE_tot
BASE_tot1['share'] = ((BASE_tot1['pop'] / BASE_tot1['World'])*100)
BASE_tot1=BASE_tot1.pivot_table(index='year',columns='country',values='share',aggfunc='sum') 
BASE_tot1
```

### Evolution of the share of the world population by countries 

```python
BASE_tot1 = BASE_tot1.dropna(axis=0, how='any') 
BASE_tot1
```

### Evolution of the share of the world population by region

```python
from pandas_datareader import data, wb

#Data from the WB
wbcountries = wb.get_countries()
wbcountries['name'] = wbcountries.name.str.strip()
```

```python
wbcountries.rename(columns={'iso3c':'countrycode'}, inplace=True) #Rename the variable to future merges
wbcountries
```

```python
BASE_tot_regions=BASE_tot[['countrycode', 'year', 'pop', 'World']].merge(wbcountries[['countrycode','region']], on=['countrycode'])
BASE_tot_regions
```

```python
BASE_tot_regions1=BASE_tot_regions
BASE_tot_regions1['share'] = ((BASE_tot_regions1['pop'] / BASE_tot_regions1['World'])*100)
BASE_tot_regions1=BASE_tot_regions1.pivot_table(index='year',columns='region',values='share',aggfunc='sum') 
BASE_tot_regions1
```

```python
BASE_tot_regions1 = BASE_tot_regions1.dropna(axis=0, how='any') 
BASE_tot_regions1
```

```python
# Plot
#plot with seaborn
sns.set(rc={'figure.figsize':(11.7,8.27)})
fig, ax = plt.subplots()
sns.lineplot( data=BASE_tot_regions1, alpha=1,  style='country', dashes=False, markers=True, linewidth=2, markersize=6)
ax.tick_params(axis = 'both', which = 'major', labelsize=12)
ax.tick_params(axis = 'both', which = 'minor', labelsize=12)

ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title("Evolution of the share of the world population by region", fontsize=18)
plt.savefig(pathgraphs + 'world_population_by_region.pdf', dpi=300, bbox_inches='tight')
```

## 3) Fertility, mortality and life expectancy data from the WB

```python
from pandas_datareader import data, wb
```

```python
wbcountries = wb.get_countries()
wbcountries['name'] = wbcountries.name.str.strip()
```

```python
# world bank search
popvars = wb.search(string='fertility')
popvars
```

```python
demographicfields = ['SP.DYN.TFRT.IN','SP.DYN.CDRT.IN','SP.DYN.LE00.IN' ] #Fertility rate, Death rate, Life expectancy at birth,Birth( 
WB_demographic= wb.download(indicator=demographicfields, country=wbcountries.iso2c.values, start=1960, end=2020)
WB_demographic
```

```python
WB_demographic = WB_demographic.dropna(axis=0, how='any') 
WB_demographic=WB_demographic.reset_index()
WB_demographic
```

```python
WB_demographic['year'] = WB_demographic.year.astype(int)
```

```python
#variables names
WB_demographic.rename(columns={'SP.DYN.TFRT.IN':'Fertility_rate'}, inplace=True) 
WB_demographic.rename(columns={'SP.DYN.CDRT.IN':'Death_rate'}, inplace=True) 
WB_demographic.rename(columns={'SP.DYN.LE00.IN':'Life_expectancy'}, inplace=True) 
```

```python
WB_demographic1=WB_demographic[WB_demographic.country=="World"]
WB_demographic1
```

```python
WB_demographic1=WB_demographic1.pivot_table(index='year',columns='country',aggfunc='sum') 
WB_demographic1
```

```python
WB_demographic2=WB_demographic1.drop(columns=['Life_expectancy'])

# Select some colors
mycolors = ["#9b59b6","#e74c3c"]
# Use seaborn to setup a color map to be used by matplotlib
my_cmap = mpl.colors.ListedColormap(sns.color_palette(mycolors).as_hex())

#plot with seaborn
sns.set(rc={'figure.figsize':(11.7,8.27)})
fig, ax = plt.subplots()
sns.lineplot( data=WB_demographic2, alpha=1, palette=sns.color_palette(mycolors), style='country', dashes=False, markers=True, linewidth=2, markersize=6)
ax.tick_params(axis = 'both', which = 'major', labelsize=12)
ax.tick_params(axis = 'both', which = 'minor', labelsize=12)

ax.set_xlabel('Year')
ax.set_ylabel('Rate')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title("Fertility and mortality data from the WB", fontsize=18)
plt.savefig(pathgraphs + 'demographicfields.pdf', dpi=300, bbox_inches='tight')
```

```python
WB_demographic3=WB_demographic1.drop(columns=['Death_rate', 'Fertility_rate'])

#plot with seaborn
sns.set(rc={'figure.figsize':(11.7,8.27)})
fig, ax = plt.subplots()
sns.lineplot( data=WB_demographic3, alpha=1, style='country', dashes=False, markers=True, linewidth=2, markersize=6)
ax.tick_params(axis = 'both', which = 'major', labelsize=12)
ax.tick_params(axis = 'both', which = 'minor', labelsize=12)

ax.set_xlabel('Year')
ax.set_ylabel('Years')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title("Life expectancy data from the WB", fontsize=18)
plt.savefig(pathgraphs + 'Life_expectancy.pdf', dpi=300, bbox_inches='tight')
```

## 4) Mortality and life expectancy data (across regions and cohorts) from the Human Mortality Database

```python
#Data available in https://www.mortality.org/
base_mortality = pd.read_csv(pathout + 'stmf.csv', sep=',',  skiprows=2)
print(base_mortality)
```

```python
#Base from the Worl Bank
wbcountries.rename(columns={'iso3c':'CountryCode'}, inplace=True) #Rename the variable to future merges
wbcountries
```

```python
# merge between the data from the human mortality database and de WB
base_mortality_region=base_mortality.merge(wbcountries, on=["CountryCode"])
base_mortality_region
```

```python
#data of both male and female
base_mortality_region_B=base_mortality_region.loc[base_mortality_region.Sex=='b']
base_mortality_region_B
```

```python
#Total death rate data by region
base_mortality_region_pivot=base_mortality_region_B.pivot_table(index='Year',columns='region', values='RTotal', aggfunc='sum') 
base_mortality_region_pivot
```

### Plot Total death rate data by region

```python
#plot with seaborn
mycolors = ["#3498db", "#95a5a6"]
# Use seaborn to setup a color map to be used by matplotlib
my_cmap = mpl.colors.ListedColormap(sns.color_palette(mycolors).as_hex())

sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("talk")

# Plot
fig, ax = plt.subplots()
sns.lineplot(data=base_mortality_region_pivot, alpha=1, style='country', dashes=False, markers=True,)
ax.tick_params(axis = 'both', which = 'major')
ax.tick_params(axis = 'both', which = 'minor')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
#ax.legend(title='', prop={'size': 40})
ax.set_xlabel('Year')
ax.set_ylabel('ratio of death counts')
ax.set_title("Total death rate data by region", fontsize=18)
plt.savefig(pathgraphs + 'Total_death_rate_data_by_region.pdf', dpi=300, bbox_inches='tight')
```

## 5) Analyze the convergence of life expectanty, mortality and fertility.

#### Let's start by looking at the distribution of the variables in 1960. For these we need to subset our dataframe and select only the rows for the year 1960. This is don with the loc property of the dataframe.

```python
#WB_demographic by year
WB_demographic1960 = WB_demographic.loc[WB_demographic.year==1960]
WB_demographic1970 = WB_demographic.loc[WB_demographic.year==1970]
WB_demographic1980 = WB_demographic.loc[WB_demographic.year==1980]
WB_demographic1990 = WB_demographic.loc[WB_demographic.year==1990]
WB_demographic2000 = WB_demographic.loc[WB_demographic.year==2000]
WB_demographic2010 = WB_demographic.loc[WB_demographic.year==2010]
WB_demographic2018 = WB_demographic.loc[WB_demographic.year==2018]
WB_demographic2018
```

### Fertility rate

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("paper")
# Plot
fig, ax = plt.subplots()
sns.kdeplot(WB_demographic1960.Fertility_rate, ax=ax, shade=True, label='1960', linewidth=2)
sns.kdeplot(WB_demographic1970.Fertility_rate, ax=ax, shade=True, label='1970', linewidth=2)
sns.kdeplot(WB_demographic1980.Fertility_rate, ax=ax, shade=True, label='1980', linewidth=2)
sns.kdeplot(WB_demographic1990.Fertility_rate, ax=ax, shade=True, label='1990', linewidth=2)
sns.kdeplot(WB_demographic2000.Fertility_rate, ax=ax, shade=True, label='2000', linewidth=2)
sns.kdeplot(WB_demographic2018.Fertility_rate, ax=ax, shade=True, label='2018', linewidth=2)
ax.set_xlabel('Fertility_rate')
ax.set_ylabel('Density of Countries')
ax.set_title("Fertility rate density per year", fontsize=18)
plt.savefig(pathgraphs + 'Fertility_rate_density.pdf', dpi=300, bbox_inches='tight')
```

### Mortality rate

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("paper")
# Plot
fig, ax = plt.subplots()
sns.kdeplot(WB_demographic1960.Death_rate, ax=ax, shade=True, label='1960', linewidth=2)
sns.kdeplot(WB_demographic1970.Death_rate, ax=ax, shade=True, label='1970', linewidth=2)
sns.kdeplot(WB_demographic1980.Death_rate, ax=ax, shade=True, label='1980', linewidth=2)
sns.kdeplot(WB_demographic1990.Death_rate, ax=ax, shade=True, label='1990', linewidth=2)
sns.kdeplot(WB_demographic2000.Death_rate, ax=ax, shade=True, label='2000', linewidth=2)
sns.kdeplot(WB_demographic2018.Death_rate, ax=ax, shade=True, label='2018', linewidth=2)
ax.set_xlabel('Death_rate')
ax.set_ylabel('Density of Countries')
ax.set_title("Death rate density per year", fontsize=18)
plt.savefig(pathgraphs + 'Death_rate_density.pdf', dpi=300, bbox_inches='tight')
```

### Life expectancy

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("paper")
# Plot
fig, ax = plt.subplots()
sns.kdeplot(WB_demographic1960.Life_expectancy, ax=ax, shade=True, label='1960', linewidth=2)
sns.kdeplot(WB_demographic1970.Life_expectancy, ax=ax, shade=True, label='1970', linewidth=2)
sns.kdeplot(WB_demographic1980.Life_expectancy, ax=ax, shade=True, label='1980', linewidth=2)
sns.kdeplot(WB_demographic1990.Life_expectancy, ax=ax, shade=True, label='1990', linewidth=2)
sns.kdeplot(WB_demographic2000.Life_expectancy, ax=ax, shade=True, label='2000', linewidth=2)
sns.kdeplot(WB_demographic2018.Life_expectancy, ax=ax, shade=True, label='2018', linewidth=2)
ax.set_xlabel('Life_expectancy')
ax.set_ylabel('Density of Countries')
ax.set_title("Life expectancy density per year", fontsize=18)
plt.savefig(pathgraphs + 'Life_expectancy.pdf', dpi=300, bbox_inches='tight')
```
